```python
import pandas as pd
import spacy
import os
```

```python
nlp1 = spacy.load(r"./output/model-best")
```

```python
full_train_tsv_filename = os.path.join("full_raw_data", "full_privatized_dataset.tsv")
train_df = pd.read_csv(full_train_tsv_filename, sep='\t')
```

```python
labels = [
    'EFFECT',
    'DOSAGE',
    'PREVIOUS CONDITION',
    'CURRENT CONDITION',
    'PRIVATE INFORMATION',
    'DURATION',
    'RECOMMENDATION',
    'DRUG NAME'
    ]

plural_labels = {
    'EFFECT': 'EFFECTS',
    'DOSAGE': 'DOSAGES',
    'PREVIOUS CONDITION': 'PREVIOUS CONDITIONS',
    'CURRENT CONDITION': 'CURRENT CONDITIONS',
    'PRIVATE INFORMATION': 'PRIVATE INFORMATION',
    'DURATION': 'DURATIONS',
    'RECOMMENDATION': 'RECOMMENDATIONS',
    'DRUG NAME': 'DRUG NAMES'
}

def labels_dict_to_text(labels_dict):

    res = ''

    for label in labels_dict:

        if not labels_dict[label]:
            continue

        num_phrases = len(labels_dict[label])

        if num_phrases > 1:
            sent = plural_labels[label].title() + ' are '
        else:
            sent = label.title() + ' is '

        

        for i, phrase in enumerate(labels_dict[label]):
            phrase = str(phrase)

            if phrase[-1] == '.':
                phrase = phrase[:-1]

            if i == (num_phrases-1):
                sent = sent + phrase + "."
            else:
                sent = sent + phrase + ", "
        
        
        res = res + sent + '\n'
    
    return res[:-1]
```

```python

train_data_dict = {
    "benefitsReview": [],
    "sideEffectsReview": [],
    "commentsReview": []
}

for col in ("benefitsReview", "sideEffectsReview", "commentsReview"):

    for text in train_df[col]:

        if not isinstance(text, str):
            print(type(text))
            print(text)
            
            train_data_dict[col].append("")
            continue

        labels_dict = {k:[] for k in labels}

        doc = nlp1(text)

        for ent in doc.ents:
            labels_dict[ent.label_].append(ent)
        
        labels_text = labels_dict_to_text(labels_dict)

        train_data_dict[col].append(labels_text)

labeled_train_df = pd.DataFrame(data=train_data_dict)
```

```python
output_filepath = os.path.join("full_labeled_output_data", "full_privatized_data_output.csv")
labeled_train_df.to_csv(output_filepath)
```
